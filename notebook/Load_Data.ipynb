```python
df_sls_all = {}

for kab in sheet_kabkota:
    df = pd.read_excel(file_path, sheet_name=kab, dtype=str)

    for c, n in [                                                                 # padding code
        ("kdkab", 2),
        ("kdkec", 3),
        ("kddesa", 3),
        ("kdsls", 4),
        ("kdsubsls", 2),
    ]:
        if c in df.columns:
            df[c] = df[c].astype(str).str.zfill(n)

    df["_row_order"] = range(len(df))                                             # save the original row order

    df_sls_all[kab] = df                                                          # save directly without changing usaha_rev
```
